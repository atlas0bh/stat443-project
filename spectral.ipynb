Here we will do a bit of spectral methods to show the seasonal nature of the data.


```r
fdf <- read.csv("data/fluview_clean/ilinet_final.csv")

data = as.numeric(fdf$age_0_4)
n = length(data) 
tt = 1:n
for(j in 1:(n/2)){ 
    om = 2*pi*j/n
    x1 = cos(om*tt)
    x2 = sin(om*tt)
    fit = lm(data~x1+x2)
    summ = summary(fit)
    if(summ$adj.r.squared>0.05){
        cat("\nj=",j,"\n"); print(summary(fit))
    }
}
```
